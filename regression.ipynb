## Structure de la rréseau de neurone utilisé

**Couche d'entré**: 81 neurones  
**couche caché 1**: 200 neurones

### Matrice des données

$X=(1000,81)$

### Matrice poids

$ W^{[1]} = (81,200) $  
$ W^{[2]} = (200,300) $  
$ W^{[3]} = (300,200) $  
$ W^{[4]} = (200,100) $  
$ W^{[5]} = (100,1) $  

### Matrice biais

$ B^{[1]} = (200,1) $  
$ B^{[2]} = (300,1) $  
$ B^{[3]} = (200,1) $  
$ B^{[4]} = (100,1) $  
$ B^{[5]} = (1,1) $ 

### Fonction d'activation utilisé dans chaque couche

Dans le couche caché, on utilise $ f(x)=max(0,x) $ et dans la couche sortie $ f(x) = x$

### Fonction coût

on utilise la fonction erreur quadratique moyenne **(Mean Squared Error)** $MSE= \frac{1}{n} \sum (y_i-s_i )^2$

### Propagation vers l'avant

$$ Z1= W^{[1]}.X + B^{[1]}$$
$$ S1=max(0,Z1) $$

$$ Z2= W^{[2]}.S1 + B^{[2]}$$
$$ S2=max(0,Z2) $$

$$ Z3= W^{[3]}.S2 + B^{[3]}$$
$$ S3=max(0,Z3) $$

$$ Z4= W^{[4]}.S3 + B^{[4]}$$
$$ S4=max(0,Z4) $$

$$ Z5= W^{[5]}.S4 + B^{[5]}$$
$$ S5=Z5 $$

### Calcul dérivée de la fonction de coût par rapport à chaque poids

Dérivée $MSE$ par rapport à $ W^{[5]}$ et  $ B^{[5]}$

$$(1) =>\frac{\partial (MSE)}{\partial W^{[5]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial W^{[5]}}$$

$$(1)^{'} =>\frac{\partial (MSE)}{\partial B^{[5]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial B^{[5]}}$$

$$ (3)^{'}=>\frac{\partial (MSE)}{\partial B^{[3]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}} .\frac{\partial Z^{[4]}}{\partial S^{[3]}}.\frac{\partial S^{[3]}}{\partial Z^{[3]}}.\frac{\partial Z^{[3]}}{\partial B^{[3]}}$$

Dérivée $MSE$ par rapport à $ W^{[2]}$ et  $ B^{[2]}$

$$ (4)=>\frac{\partial (MSE)}{\partial W^{[2]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}} .\frac{\partial Z^{[4]}}{\partial S^{[3]}}.\frac{\partial S^{[3]}}{\partial Z^{[3]}}.\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial W^{[2]}}$$
$$ (4)^{'}=>\frac{\partial (MSE)}{\partial B^{[2]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}} .\frac{\partial Z^{[4]}}{\partial S^{[3]}}.\frac{\partial S^{[3]}}{\partial Z^{[3]}}.\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial B^{[2]}}$$

Dérivée $MSE$ par rapport à $ W^{[1]}$ et  $ B^{[1]}$

$$ (5)=>\frac{\partial (MSE)}{\partial W^{[1]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}} .\frac{\partial Z^{[4]}}{\partial S^{[3]}}.\frac{\partial S^{[3]}}{\partial Z^{[3]}}.\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial S^{[1]}}.\frac{\partial S^{[1]}}{\partial Z^{[1]}}.\frac{\partial Z^{[1]}}{\partial W^{[1]}}$$

$$ (5)^{'}=>\frac{\partial (MSE)}{\partial B^{[1]} } = \frac{\partial (MSE)}{\partial S^{[5]}} . \frac{\partial S^{[5]}}{\partial Z^{[5]}}. \frac{\partial Z^{[5]}}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}} .\frac{\partial Z^{[4]}}{\partial S^{[3]}}.\frac{\partial S^{[3]}}{\partial Z^{[3]}}.\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial S^{[1]}}.\frac{\partial S^{[1]}}{\partial Z^{[1]}}.\frac{\partial Z^{[1]}}{\partial B^{[1]}}$$

**Matrice des poids**

```python
# Initialisation des matrices de poids et de biais
# Dimensions: (nombre de neurones dans la couche suivante, nombre de neurones dans la couche actuelle)
np.random.seed(42)

# Initialisation des poids et des biais
W1 = np.random.randn(200, 81) * 0.01
W2 = np.random.randn(300, 200) * 0.01
W3 = np.random.randn(200, 300) * 0.01
W4 = np.random.randn(100, 200) * 0.01
W5 = np.random.randn(1, 100) * 0.01

B1 = np.zeros((200, 1))
B2 = np.zeros((300, 1))
B3 = np.zeros((200, 1))
B4 = np.zeros((100, 1))
B5 = np.zeros((1, 1))
```

**Fonction utilisé**

```python
# Fonction ReLU
def reLU(Z):
    return np.where(Z < 0, 0, Z)

# Dérivée de la fonction ReLU
def reLuDerivate(Z):
    return np.where(Z < 0, 0, 1)

# Propagation linéaire
def linear(W, S, B):
    return np.dot(W, S) + B
```

**Propagation vers l'avant**

```python

# Propagation vers l'avant
def forward_propagation(X, W1, W2, W3, W4, W5, B1, B2, B3, B4, B5):
    Z1 = linear(W1, X.T, B1)
    S1 = reLU(Z1)

    Z2 = linear(W2, S1, B2)
    S2 = reLU(Z2)
    
    Z3 = linear(W3, S2, B3)
    S3 = reLU(Z3)
    
    Z4 = linear(W4, S3, B4)
    S4 = reLU(Z4)

    Z5 = linear(W5, S4, B5)
    S5 = Z5
    
    return S1, S2, S3, S4, S5
```

**Fonction coût**

```python
# Fonction coût mse
def mse(S, Y):
    n = Y.shape[1]
    return (1 / n) * np.sum((Y - S)**2)
```

**Propagation vers l'arrière**

```python
# Calcul des dérivées pour la descente de gradient
def back_propagation(S1, S2, S3, S4, S5, W1, W2, W3, W4, W5, X, Y):
    m = Y.shape[1]  # Nombre d'exemples

    d5 = S5 - Y
    DW5 = (2 / m) * np.dot(d5, S4.T)
    DB5 = (2 / m) * np.sum(d5, axis=1, keepdims=True)

    d4 = np.dot(W5.T, d5) * reLuDerivate(S4)
    DW4 = (2 / m) * np.dot(d4, S3.T)
    DB4 = (2 / m) * np.sum(d4, axis=1, keepdims=True)

    d3 = np.dot(W4.T, d4) * reLuDerivate(S3)
    DW3 = (2 / m) * np.dot(d3, S2.T)
    DB3 = (2 / m) * np.sum(d3, axis=1, keepdims=True)

    d2 = np.dot(W3.T, d3) * reLuDerivate(S2)
    DW2 = (2 / m) * np.dot(d2, S1.T)
    DB2 = (2 / m) * np.sum(d2, axis=1, keepdims=True)

    d1 = np.dot(W2.T, d2) * reLuDerivate(S1)
    DW1 = (2 / m) * np.dot(d1, X)
    DB1 = (2 / m) * np.sum(d1, axis=1, keepdims=True)

    return DW1, DB1, DW2, DB2, DW3, DB3, DW4, DB4, DW5, DB5

  
```

**Propagation vers l'arrière**

```python
 #Mise à jour des paramètres
def update_parameters(X, Y, W1, B1, W2, B2, W3, B3, W4, B4, W5, B5, learning_rate=0.01, n=10):
    data_cost = []
    for i in range(n):
        S1, S2, S3, S4, S5 = forward_propagation(X, W1, W2, W3, W4, W5, B1, B2, B3, B4, B5)
        
        # Calcul de la fonction coût
        cost = mse(S5, Y)
        data_cost.append(cost)
        
        print(f"Iteration {i+1}/{n}, mse: {float(cost)}")
        
        # Propagation vers l'arrière
        DW1, DB1, DW2, DB2, DW3, DB3, DW4, DB4, DW5, DB5 = back_propagation(S1, S2, S3, S4, S5, W1, W2, W3, W4, W5, X, Y)

        # Mise à jour des paramètres
        W1 -= learning_rate * DW1
        B1 -= learning_rate * DB1

        W2 -= learning_rate * DW2
        B2 -= learning_rate * DB2

        W3 -= learning_rate * DW3
        B3 -= learning_rate * DB3

        W4 -= learning_rate * DW4
        B4 -= learning_rate * DB4

        W5 -= learning_rate * DW5
        B5 -= learning_rate * DB5
    
    return W1, B1, W2, B2, W3, B3, W4, B4, W5, B5, data_cost, S5
```

```python
# Initialisation des poids et des biais
W1 = np.random.randn(200, 81) * 0.01
W2 = np.random.randn(300, 200) * 0.01
W3 = np.random.randn(200, 300) * 0.01
W4 = np.random.randn(100, 200) * 0.01
W5 = np.random.randn(1, 100) * 0.01

B1 = np.zeros((200, 1))
B2 = np.zeros((300, 1))
B3 = np.zeros((200, 1))
B4 = np.zeros((100, 1))
B5 = np.zeros((1, 1))
S1, S2, S3, S4, S5 = forward_propagation(X, W1, W2, W3, W4, W5, B1, B2, B3, B4, B5)
DW1, DB1, DW2, DB2, DW3, DB3, DW4, DB4, DW5, DB5 = back_propagation(S1, S2, S3, S4, S5, W1, W2, W3, W4, W5, X, Y)

W1, B1, W2, B2, W3, B3, W4, B4, W5, B5, data_cost, y_pred = update_parameters(X, Y, W1, B1, W2, B2, W3, B3, W4, B4, W5, B5, 0.01, 1000)
```
