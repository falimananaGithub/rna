### II-4 Propagation vers l'arrière

#### Fonction coût

Pour appliquer la propagation vers l'arrière, il faut définir la fonction de coût.
Une fonction de coût (Loss Function) est une fonction qui permet de mesurer les erreurs effectuées par un modèle.
Les erreurs sont les distances entre la sortie générée par le modèle et la sortie désirée.
La fonction sigmoïde est une fonction logistique qui renvoie une probabilité.

Si on considère que la probabilité qu'un grain de riz appartient à la variété Cammeo est égale à p, alors celle qu'il appartient à la variété Ommincik est égale à 1-p.

$$ P(Y=y)=p^{y}(1-p)^{1-y} $$

**Vraisemblance**

En statistique, la vraisemblance est une mesure de la compatibilité d'un ensemble de données observées avec un modèle statistique donné. Elle est utilisée pour estimer les paramètres du modèle.
Ici, La fonction vraisemblance du modèle est le produit de toute probabilité de données. 
$$ L =  \prod_{i=1}^{n} P(Y=y_i)= \prod_{i=1}^{n} s^{y_i}_i(1-s_i)^{1-y_i}$$
$$y_i: \text{sortie  desiré} \text{ , }s_i: \text{sortie genéré} $$

Si le resultat de **L** (vraisemblance) est proche de 1 alors on dit que le resultat est vraisemblable à 100 % 

Mais le problème de vraisemblance c'est que plus il y a de nombres, plus le resultat tend vers 0. 
c'est pour cela qu'on doit appliquer la fonction logarithme dans cette fonction car: $$log(xy)=log(x)+log(y)$$


La fonction vraisemblance devient comme suit: $$ log(L)=log(\prod_{i=1}^{n} s^{y_i}_i(1-s_i)^{1-y_i}) $$

Ici,la log-vraisemblance est concave (ou convexe) car c'est une régression logistique. Cela signifie qu'il existe une seule valeur maximale (un maximum global) pour la log-vraisemblance, facilitant les méthodes de maximisation comme la descente de gradient.

On sait que 
$$
\begin{flalign}
 Argmax(L) = Argmax(Log(L))
\end{flalign}
$$

alors on peut prendre la fonction log-vraisemblance pour continuer le caclul

$$ LL = log(L)=log(\prod_{i=1}^{n} s^{y_i}_i(1-s_i)^{1-y_i}) $$

$$=\sum_{i=1}^{n}log( s^{y_i}_i(1-s_i)^{1-y_i}) $$

$$ =\sum_{i=1}^{n}log( s^{y_i}_i)+log((1-s_i)^{1-y_i})$$

$$ LL =\sum_{i=1}^{n} y_ilog( s^{[4]}_i)+(1-y_i)log((1-s^{[4]}_i))$$

A partir cette fonction, on obtient la fonction coût qui s'appelle fonction **"Log Loss"**

$$ \text{Log Loss} =- \frac{1}{m} \sum_{i=1}^{m} y_ilog( s^{[4]}_i)+(1-y_i)log((1-s^{[4]}_i))$$

$$ \text{Log Loss} =- \frac{1}{m} \sum_{i=1}^{m} y_ilog( S^{[4]})+(1-y_i)log((1-S^{[4]}))$$


Pour déterminer le meilleur modèle, on doit maximiser la fonction de log-vraisemblance.   Cependant, en général, il n'existe pas d'algorithme universel pour maximiser une fonction.  
C'est pourquoi le signe **'-'** est utilisé, car la maximisation d'une fonction $f(x)$ est équivalente à la minimisation de la fonction $−f(x)$. 

Et le $ \frac{1}{m} $ est juste normaliser le resultat.

**Descente de Gradient**

L'algoritheme descente de gradient est utiliser pour déterminer la minimisation de la fonction Log Loss. c'est la dérivée de la fonction Log Loss par rapport au poids.
Cet algorithme peut déterminer la valeur des poids et biais pour que le modèle soit le meilleur.
$$ W^{[n]}_{i+1}=W^{[n]}_{i}-\eta \frac{\partial (LL)}{\partial W^{[n]}} $$
$$ B^{[n]}_{i+1}=B^{[n]}_{i}-\eta \frac{\partial (LL)}{\partial B^{[n]}} $$

**Calcule de gradient**

 $$ (1)=> \frac{\partial (LL)}{\partial W^{[4]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial W^{[4]}}$$

$$ (1)^{'}=>\frac{\partial (LL)}{\partial B^{[4]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial B^{[4]}}$$

 $$ (2)=> \frac{\partial (LL)}{\partial W^{[3]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}} .\frac{\partial Z^{[3]}}{\partial W^{[3]}}$$

 $$ (2)^{'}=>\frac{\partial (LL)}{\partial B^{[3]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}} .\frac{\partial Z^{[3]}}{\partial B^{[3]}}$$

$$ (3)=>\frac{\partial (LL)}{\partial W^{[2]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}} .\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial W^{[2]}}$$

$$ (3)^{'}=>\frac{\partial (LL)}{\partial B^{[2]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}} .\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial B^{[2]}}$$

$$ (4)=> \frac{\partial (LL)}{\partial W^{[1]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}} .\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial S^{[1]}} . \frac{\partial S^{[1]}}{\partial Z^{[1]}} .\frac{\partial Z^{[1]}}{\partial W^{[1]}}$$

$$ (4)^{'}=>\frac{\partial (LL)}{\partial B^{[1]} } = \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}. \frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}} .\frac{\partial Z^{[3]}}{\partial S^{[2]}}.\frac{\partial S^{[2]}}{\partial Z^{[2]}}.\frac{\partial Z^{[2]}}{\partial S^{[1]}} . \frac{\partial S^{[1]}}{\partial Z^{[1]}} .\frac{\partial Z^{[1]}}{\partial B^{[1]}}$$

**Calcul de l'equation $(1)$ et $ (1)^{'}$**

$ (1) =  \frac{1}{m} d^{[4]}.\frac{\partial Z^{[4]}}{\partial W^{[4]}} $  où  $ d^{[4]}= \frac{\partial (LL)}{\partial S^{[4]}} . \frac{\partial S^{[4]}}{\partial Z^{[4]}}= (S^{[4]} - Y)$ alors on a $ (1) =   \frac{1}{m} d^{[4]}.S^{[3]} $  car $ \frac{\partial Z^{[4]}}{\partial W^{[4]}} = S^{[3]}$

$ (1)^{'} = \frac{1}{m} \sum_{s.e_1} d^{[4]}.\frac{\partial Z^{[4]}}{\partial B^{[4]}} $  alors on a $ (1)^{'} =  \frac{1}{m} \sum_{s.e_1}(S^{[4]} - Y) $  car $ \frac{\partial Z^{[4]}}{\partial B^{[4]}} = 1$  

En resumé:
$$ \frac{\partial (LL)}{\partial W^{[4]} } =  \frac{1}{m} d^{[4]}.S^{[3]}$$
$$ \frac{\partial (LL)}{\partial B^{[4]} } =  \frac{1}{m} \sum_{s.e_1} d^{[4]} $$

**Calcul de l'equation $(2)$ et $ (2)^{'}$**

$ (2) = \frac{1}{m} d^{[3]}.\frac{\partial Z^{[3]}}{\partial W^{[3]}} $  où  $ d^{[3]}= d^{[4]} .\frac{\partial Z^{[4]}}{\partial S^{[3]}} . \frac{\partial S^{[3]}}{\partial Z^{[3]}}= d^{[4]}.W^{[4]}.S^{[3]}(1-S^{[3]})$ alors on a $ (2) = \frac{1}{m}d^{[4]}.W^{[4]}.S^{[3]}(1-S^{[3]}).S^{[2]}$ car $\frac{\partial Z^{[3]}}{\partial W^{[3]}}=S^{[2]} $

$ (2)^{'}= \frac{1}{m} \sum_{s.e_1} d^{[3]} $

En resumé:

$$ \frac{\partial (LL)}{\partial W^{[3]} } = \frac{1}{m} d^{[3]}.S^{[2]}$$
$$ \frac{\partial (LL)}{\partial B^{[3]} } = \frac{1}{m} \sum_{s.e_1} d^{[3]}$$

**Calcul de l'equation $(3)$ et $ (3)^{'}$**

$ (3) = \frac{1}{m} d^{[2]}.\frac{\partial Z^{[2]}}{\partial W^{[2]}} $  où  $ d^{[2]}= d^{[3]} .\frac{\partial Z^{[3]}}{\partial S^{[2]}} . \frac{\partial S^{[2]}}{\partial Z^{[2]}}= d^{[3]}.W^{[3]}.S^{[2]}(1-S^{[2]})$ alors on a $ (3) = \frac{1}{m} d^{[3]}.W^{[3]}.S^{[2]}(1-S^{[2]}).S^{[1]}$ car $\frac{\partial Z^{[2]}}{\partial W^{[2]}}=S^{[1]} $


$ (3)^{'}= \frac{1}{m} \sum_{s.e_1} d^{[2]} $

En resumé:
$$ \frac{\partial (LL)}{\partial W^{[2]} } = \frac{1}{m} d^{[2]}.S^{[1]} $$
$$ \frac{\partial (LL)}{\partial B^{[2]} } = \frac{1}{m} \sum_{s.e_1} d^{[2]}  $$

**Calcul de l'equation $(4)$ et $ (4)^{'}$**

$ (4) = \frac{1}{m} d^{[1]}.\frac{\partial Z^{[1]}}{\partial W^{[1]}} $  où  $ d^{[1]}= d^{[2]} .\frac{\partial Z^{[2]}}{\partial S^{[1]}} . \frac{\partial S^{[1]}}{\partial Z^{[1]}}= d^{[2]}.W^{[2]}.S^{[1]}(1-S^{[1]})$ alors on a $ (3) = \frac{1}{m} d^{[2]}.W^{[2]}.S^{[1]}(1-S^{[1]}).X$ car $\frac{\partial Z^{[2]}}{\partial W^{[2]}}=X $

$ (4)^{'}= \frac{1}{m} \sum_{s.e_1} d^{[1]} $

En resumé:
$$ \frac{\partial (LL)}{\partial W^{[1]} } = \frac{1}{m}  d^{[1]}. X $$

### III- 2  Propagation vers l'avant

Fonction sigmoide $ S(Z) = \frac{1}{1+e^{-Z}}$

```python
# create sigmoid function
def sigmoid(Z):
    return (1/(1+np.exp(-Z)))
```

```python
def linear(W,S,B):
    if B.ndim == 2 and B.shape[0] == 1:
        B = B.flatten() 
    return (np.dot(W,S) + B)
```

 Fonction coût Log Loss $$ \text{Log Loss} =- \frac{1}{m} \sum_{i=1}^{m} y_ilog( S^{[4]})+(1-y_i)log((1-S^{[4]}))$$

```python
def log_loss(S,Y):
    # Convertir en tableaux NumPy si ce sont des pandas Series ou DataFrame
    if isinstance(S, (pd.Series, pd.DataFrame)):
        S = S.values
    if isinstance(Y, (pd.Series, pd.DataFrame)):
        Y = Y.values

    # Nombre total d'exemples
    m = len(Y)
    
    if S.shape != Y.shape:
        raise ValueError("Les dimensions de S et Y doivent correspondre.")
    
    # Clip les valeurs pour éviter log(0) ou log(1)
    S = np.clip(S, 1e-15, 1 - 1e-15)
    LL = - (1 / m) * np.sum(Y * np.log(S) + (1 - Y) * np.log(1 - S))
    return LL
```

### III- 3  Propagation vers l'arrière

```python
def d4(S,Y):
    return (S-Y)
# cacule derivée partielle LL par rapport à W4
d_4=d4(S4.T,Y)
DW4=(1/len(Y))*np.dot(S3,d_4)

# calcule derivée partielle LL par rapport à B4
DB4=(1/len(Y))*np.sum(d4(S4.T,Y))
```

$$ \frac{\partial (LL)}{\partial W^{[3]} } = \frac{1}{m} d^{[3]}.S^{[2] T}$$
$$ \frac{\partial (LL)}{\partial B^{[3]} } = \frac{1}{m} \sum_{s.e_1} d^{[3]}$$

```python
def d3(d4,W4,S3):
    A=np.dot(d4,W4)
    B=np.dot(S3.T,(1-S3))
    return (np.dot(A.T,B))
    
d_3=d3(d_4,W4.T,S3)

DW3=1/len(Y)*np.dot(d_3,S2.T)

DB3=(1/len(Y))*np.sum(d_3)

print(DB3)
```

Calcul derivée de la fonction Log Loss par rapport à $ W^{[2]} $ et à $B^{[2]}$
$$ d^{[2]} = d^{[3]T}.W^{[3]}.S^{[2]}(1-S^{[2]})$$
$$ \frac{\partial (LL)}{\partial W^{[2]} } = \frac{1}{m} d^{[2]}.S^{[1]} $$
$$ \frac{\partial (LL)}{\partial B^{[2]} } = \frac{1}{m} \sum_{s.e_1} d^{[2]}  $$

```python
def d2(d_3,W3,S2):
    A=np.dot(d_3.T,W3)
    B=np.dot(S2.T,(1-S2))
    return (np.dot(A.T,B))
    
d_2=d2(d_3,W3.T,S2)

DW2=1/len(Y)*np.dot(d_2,S1.T)

DB2=(1/len(Y))*np.sum(d_2)

print(DB2)
```

$$ d^{[1]} = d^{[2]}.W^{[2]}.S^{[1]}(1-S^{[1]})$$
$$ \frac{\partial (LL)}{\partial W^{[1]} } = \frac{1}{m}  d^{[1]}. X $$
$$ \frac{\partial (LL)}{\partial B^{[1]} } = \frac{1}{m} \sum_{s.e_1} d^{[1]} $$

```python
def gradient(S1,S2,S3,S4,W1,W2,W3,W4,X,Y):
    
    d_4=d4(S4.T,Y)
    DW4=(1/len(Y))*np.dot(S3,d_4)
    DB4=(1/len(Y))*np.sum(d4(S4.T,Y))

    d_3=d3(d_4,W4.T,S3)
    DW3=1/len(Y)*np.dot(d_3,S2.T)
    DB3=(1/len(Y))*np.sum(d_3)

    d_2=d2(d_3,W3.T,S2)
    DW2=1/len(Y)*np.dot(d_2,S1.T)
    DB2=(1/len(Y))*np.sum(d_2)

    d_1=d1(d_2,W2.T,S1)
    DW1=1/len(Y)*np.dot(d_1,X)
    DB2=(1/len(Y))*np.sum(d_1)

    return (DW4,DW3,DW2,DW1,DB4,DB3,DB2,DB1)
    

def forward_propagation(W1,W2,W3,W4,B1,B2,B3,B4,X,Y):
    
    Z1=linear(W1,X.T,B1)
    S1=sigmoid(Z1)

    Z2=linear(W2,S1,B2)
    S2=sigmoid(Z2)

    Z3=linear(W3,S2,B3)
    S3=sigmoid(Z3)
    
    Z4 = linear(W4.T,S3,B4)
    S4=sigmoid(Z4)
    
    DW4,DW3,DW2,DW1,DB4,DB3,DB2,DB1=gradient(S1,S2,S3,S4,W1,W2,W3,W4,X,Y)
    

    return (DW4,DW3,DW2,DW1,DB4,DB3,DB2,DB1,log_loss(S4.T,Y))


        
def train(n,alpha,W1,W2,W3,W4,B1,B2,B3,B4,DW4,DW3,DW2,DW1,DB4,DB3,DB2,DB1,X,Y):
    data_error=[]

    for i in range(0,n+1):
        #mis à jours des poids par descente de gradient
        W1=W1-alpha*DW1
        W2=W2-alpha*DW2
        W3=W3.T-alpha*DW3
        W4=W4-alpha*DW4

        #mis à jours de biais
        B1=B1-alpha*DB1
        B2=B2-alpha*DB2
        B3=B3-alpha*DB3
        B4=B4-alpha*DB4

        DW4,DW3,DW2,DW1,DB4,DB3,DB2,DB1,log_loss = forward_propagation(W1,W2,W3,W4,B1,B2,B3,B4,X,Y)
        data_error.append(log_loss)
    return data_error

train(10,0.01,W1,W2,W3,W4,B1,B2,B3,B4,DW4,DW3,DW2,DW1,DB4,DB3,DB2,DB1,X,Y)

        

        
        
    
```
